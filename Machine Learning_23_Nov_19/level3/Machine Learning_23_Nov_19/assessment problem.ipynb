```python
import pandas as pd
bird = pd.read_csv("bird.csv")
bird.sample(5)
```

```python
x = bird.drop(columns=["id","type"],axis=1)
y = bird[["type"]]
x
```

```python
import numpy as np
bird = bird.replace(np.NaN,0)
bird.sample(10)
```

```python
bird.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_y = LabelEncoder()
bird["type"] = labelencoder_y.fit_transform(y)
# print(y)
bird

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 0)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel= 'linear',random_state=0) # kernel can be polynomial, RBF
print(classifier.fit(x_train,y_train))
```

```python
y_predict=classifier.predict(x_test)
y_predict
```

```python
#Evaluating Performance
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report,confusion_matrix
print("==confusion Matrix==>")
conff=confusion_matrix(y_test,y_predict)
print(conff)
print("\n")
c=print(conff[0,1]+conff[1,0])#FP(false+ve) and FN(False-ve)
print(conff[0,0]+conff[1,1])#True +ve and True _ve
from sklearn.metrics import accuracy_score
print("Acuuracy:",accuracy_score(y_test,y_predict)*100)
```

# decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
x_bird = DecisionTreeRegressor(max_depth = 3)
x_bird.fit(x,y)
```

```python
y_pred = x_bird.predict(x_test)
y_pred
```
