# Simple Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.random.rand(100,1)
x
```

```python
y=4+3*x+np.random.rand(100,1)
y
```

```python
plt.plot(x,y,'.',color="r")
```

```python
x_b=np.c_[np.ones((100,1)),x] #adding x0=1 to each instance adding ones and random variables 
x_b
```

```python
theta_best=np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y)
```

```python
theta_best
```

```python
ypredicted=x_b.dot(theta_best)
ypredicted
```

```python
plt.plot(x,y,'x',label="Befor Pridiction")
plt.plot(x,ypredicted,'o',label="After Pridiction")
plt.legend()
plt.xlabel('X')
plt.ylabel('y')
plt.title("Visualisation of normal and prediction values")
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x,y)
reg.intercept_,reg.coef_
```

```python
eta = 0.1 # learning rate
n_iterations = 1000
m =100
theta = np.random.randn(2,1)
```

```python
for iteration in range(n_iterations):
    gradients = 2/m * x_b.T.dot(x_b.dot(theta)-y) # the way how to decreses the cost function is gradient 
    theta = theta - eta*gradients
```
