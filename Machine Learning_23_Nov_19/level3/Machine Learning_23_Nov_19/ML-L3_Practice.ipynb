# *Data Retrieval*

* Retrive Index columns -- df.index
* Columns names    -- df.columns
* second columns   -- df[Column_name]  # output in series format  
* any columns     -- df[[Column_name]] # output in DataFrame
* First Row      -- df.head(1)
* Access single row -- df.iloc[index]
* Access single cell

```python
import pandas as pd

df = pd.read_csv('olympics.csv', index_col=0, skiprows=1)

for col in df.columns:
    if col[:2]=='01':
        df.rename(columns={col:'Gold'+col[4:]}, inplace=True)
    if col[:2]=='02':
        df.rename(columns={col:'Silver'+col[4:]}, inplace=True)
    if col[:2]=='03':
        df.rename(columns={col:'Bronze'+col[4:]}, inplace=True)
    if col[:1]=='№':
        df.rename(columns={col:'#'+col[1:]}, inplace=True)

names_ids = df.index.str.split('\s\(') # split the index by '('

df.index = names_ids.str[0] # the [0] element is the country name (new index) 
df['ID'] = names_ids.str[1].str[:3] # the [1] element is the abbreviation or ID (take first 3 characters from that)

df = df.drop('Totals')
df.head()
```

# Pandas Topics
 * Convert Dict to DataFrames -- pd.DtaFrame({...})
 * skiprows  -- skiprows = value
 * Get top rows from df -- df.head(10)
 * Get bottom rows from df -- df.tail(10)
 * Get specified rows from df -- df.iloc[index]  or -- df.iloc[[index_val1,index_val2]]
 * Get specified row and columns -- df.loc[[Row_names],[Column_names]]
 * Get Max of column -- df[column_name].max()
 * Get attributes of Column -- df[column_name].min() df[column_name].mean()  df[column_name].median() 
 * set row as index by using --  usecol= value
 * Get Columns -- df.columns
 * Get columns slicing -- df.columns[:]
 * rename Column names  -- pd.rename(columns= { "old column name" = "new Column name"})
 * Get specified Columns in DataFrame -- df[[ column_name1,Column_name2]]
 * Get column in Series -- pd.Series(df[column_name])
 * To remove columns -- pd.drop(colum_name,axis=1,inplace=True)
 * To Convert Columns as Index -- pd.set_index(column_name)
 * To reset Index into Columns -- pd.reset_index()
 * To get index of specified value in a column -- df[column_name].idxmax() or -- df[Column_name].argmax()
 * Replace anything -- df.replace(old_val,new_val)
 * Merge of 2 files  -- pd.merge(file1,file2,how=inner/outer,right_on=Column_name,left_on=Column_name)
