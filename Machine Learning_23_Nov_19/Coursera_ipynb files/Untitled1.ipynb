```python
import pandas as pd
import numpy as np
def answer_one():
    data_file=pd.read_excel("Energy+Indicators.xls",skiprows=17,skipfooter=38,usecols=[2,3,4,5])# read excel file,del 15rows,38rows from tail,extract wanted columns
    data_file.columns=['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
    data_file=data_file.replace("...",np.NaN) # replace ... into NaN
    data_file["Energy Supply"]=data_file["Energy Supply"]*1000000  # multiplex 1000000 to energy supply
    data=data_file['Country'].str.replace(r" \(.*\)","")
    data_file["Country"]=data.str.split(r"\d{1,3}").str.get(0)
    data_file["Country"]=data_file["Country"].replace({"Republic of Korea": "South Korea","United States of America": "United States","United Kingdom of Great Britain and Northern Ireland": "United Kingdom","China, Hong Kong Special Administrative Region": "Hong Kong"})
    data_file=data_file.set_index("Country")
    dfile=pd.read_csv("world_bank.csv",skiprows=4)
    dfile= dfile.rename(columns = {"Country Name":"Country"})
    dfile["Country"]=dfile["Country"].replace({"Korea, Rep.": "South Korea", 
    "Iran, Islamic Rep.": "Iran",
    "Hong Kong SAR, China": "Hong Kong"})
    gdp=dfile[["Country","2006","2007",'2008','2009', '2010', '2011', '2012', '2013', '2014', '2015']]
    gdp=gdp.set_index("Country")
    d=pd.read_excel("scimagojr-3.xlsx")
    d=d[0:15]
    d=d.set_index("Country")
    res=pd.merge(d,data_file,how="inner",right_on="Country",left_on="Country")
    result=pd.merge(res,gdp,how="inner",right_on="Country",left_on="Country")
    return dfile

    
answer_one()
```
